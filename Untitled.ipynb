```python
# Question 2

num= int(input("Entrez un entier : "))
def factorial(num): 
    if num < 0: 
        print("LA FACTORIELLE D'UN NOMBRE NEGATIVE N'EXISTE PAS")

    elif num == 0: 
        return 1
        
    else: 
        fact = 1
        while(num > 1): 
            fact *= num 
            num -= 1
        return fact  

print("LA FACTORIELLE",num,"EST", factorial(num)) 
```
